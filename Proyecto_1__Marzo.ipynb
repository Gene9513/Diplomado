EL CAMBIO DE PRECIO DE LA TORTILLA ATRAVES DE LOS AÑOS

El aumento del precio de la tortilla es un problema nacional, ya que ha afectado negativamente la economía de la inmensa mayoría de las familias mexicanas, sobre todo en el de las más pobres. Por lo que siete de cada diez mexicanos creen que puede llegar a poner en riesgo la estabilidad social del país
analiza las razones de esta situación y concluye que la asignación de los recursos entre las empresas de la economía mexicana, es decir, cómo se distribuyen los factores de producción como el capital y el trabajo entre los productores, es una de las la raíces del dilema de la escasa productividad en el país.
Aunque en el último año el precio del maíz ha disminuido, el del kilo de tortilla no baja; sin embargo, esto se debe a factores como el desfase entre los ciclos de cosecha del grano, la diferencia de consumo entre regiones del país y las alzas en costos como el salario y la energía, entre otros.
En México, la tonelada del grano en Sinaloa ronda los 5 mil 300 pesos, mientras que el año pasado costaba unos 7 mil pesos.

Es importante tener en cuenta que las grandes empresas hacen compras de maíz por adelantado -unos cuatro meses-. “Apenas se están empezando a reflejar los menores precios”.
El mercado se rige por la oferta y la demanda. Sin embargo, hay otros aspectos que influyen en el precio de la tortilla.

Uno de ellos es la oferta y la competencia. En la CDMX, Estado de México y Puebla hay más tortillerías. Eso ayuda a que el precio no suba tanto; mientras que en el norte del país hay menos y en su lugar, consumen más las tortillas de harina de trigo.

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Year']
  ys = series['Price per kilogram']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Year', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Store type')):
  _plot_series(series, series_name, i)
  fig.legend(title='Store type', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Year')
_ = plt.ylabel('Price per kilogram')
```

```python
# Comencemos eliminando estas entradas 'NaN'
df_null = df.loc[df.isnull().any(axis=1)]
df_notnull = df.dropna()

print('Median of Price per Kg: ', df_notnull['Price per kilogram'].median())
print('Mean of Price per Kg: ', df_notnull['Price per kilogram'].mean())
print('Modal of Price per Kg: ', df_notnull['Price per kilogram'].mode())

df = df.fillna(df_notnull['Price per kilogram'].mean())

df
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  counted = (series['Year']
                .value_counts()
              .reset_index(name='counts')
              .rename({'index': 'Year'}, axis=1)
              .sort_values('Year', ascending=True))
  xs = counted['Year']
  ys = counted['counts']
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Year', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Store type')):
  _plot_series(series, series_name, i)
  fig.legend(title='Store type', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Year')
_ = plt.ylabel('count()')
```

```python
#Precio por Año

tortilla_average_price_per_year = df.groupby(["Year"])["Price per kilogram"].mean()
tortilla_average_price_mom_and_pop_per_year = df[df["Store type"] =="Mom and Pop Store"].groupby(["Year"])["Price per kilogram"].mean()
tortilla_average_price_retail_store_year = df[df["Store type"] =="Big Retail Store"].groupby(["Year"])["Price per kilogram"].mean()
plt.plot(tortilla_average_price_mom_and_pop_per_year,"g",label = "Mom and Pop Store")
plt.plot(tortilla_average_price_retail_store_year,"r",label = "Big Retail Store")
plt.legend()
plt.title("El precio de la Tortilla por año")
plt.xticks(tortilla_average_price_per_year.index[::2])
plt.xlabel("Año")
plt.ylabel("Precio por Kilogramo")
plt.grid()
#plt.show()
print(tortilla_average_price_mom_and_pop_per_year)
```

```python
df_1 = df.groupby(['Store type','Year','State'])['Price per kilogram'].mean().to_frame().reset_index()
```

```python
#precios por año y estado
sns.set_style("darkgrid")
g = sns.relplot(data=df_1, kind='line',
            x='Year', y='Price per kilogram', hue='Store type', style='Store type',
            col='State', col_wrap=4)
g.set(ylim=(0, 31), xlim=(2006.5, 2024.5), ylabel="Price per kilogram ($MXN)")
plt.show()
```

```python
df_1_wide = df_1.query("Year == 2018").pivot(index='State', columns='Store type', values='Price per kilogram')
df_1_wide['Indice de relacion de precios'] = df_1_wide['Mom and Pop Store'] / df_1_wide['Big Retail Store']
df_1_wide.head(33)
```
