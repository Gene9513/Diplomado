13. Menciona las diferencias entre "correlación" y "causalidad".

**R.**
La correlación describe la relación estadística entre dos variables que varían juntas en un patrón predecible, mientras que la causalidad se refiere a la relación de causa y efecto entre dos variables.
La correlación se puede medir cuantitativamente con coeficientes de correlación, mientras que determinar la causalidad requiere evidencia adicional y un análisis más riguroso.
Una alta correlación no implica causalidad, ya que las variables pueden estar asociadas debido a otras variables o factores no observados. La causalidad implica una relación de causa y efecto que puede ser demostrada mediante métodos específicos de investigación.

14. ¿Qué es Deep Learning?

**R.**
El Deep Learning (Aprendizaje Profundo) es una rama del aprendizaje automático que se basa en redes neuronales artificiales con múltiples capas de procesamiento para aprender representaciones jerárquicas de datos. Estas representaciones van desde características simples en las capas iniciales hasta características más abstractas y complejas en las capas posteriores. El término "profundo" se refiere al número de capas en la red neuronal, que puede variar desde unas pocas capas hasta cientos o incluso miles en modelos muy complejos.

19. Menciona 2 modelos de predicción y describe brevemente cómo funcionan.

**R.**
 Regresión Lineal:
 La regresión lineal es un modelo que establece una relación lineal entre una variable dependiente y una o más variables independientes. En su forma más simple, con una sola variable independiente, el modelo de regresión lineal ajusta una línea recta a los datos para predecir valores continuos de la variable dependiente. El objetivo es encontrar los coeficientes de pendiente e intercepto que minimizan la suma de los errores cuadrados entre las observaciones reales y las predichas por el modelo.

 Random Forest:
 Random Forest es un método de aprendizaje conjunto que combina múltiples árboles de decisión para hacer predicciones. En el caso de la regresión, Random Forest crea un conjunto de árboles de decisión y promedia sus predicciones para obtener la predicción final. Este enfoque de conjunto ayuda a reducir el sobreajuste y mejorar el rendimiento del modelo.

20. Menciona 2 modelos de clasificación y describe brevemente cómo funcionan.

**R.**

K-Nearest Neighbors (KNN):
Es un algoritmo no paramétrico que realiza predicciones basadas en el voto mayoritario de los K vecinos más cercanos de una instancia en el espacio de características. La métrica de distancia, como la distancia euclidiana, se utiliza para determinar la proximidad entre instancias. KNN no aprende parámetros explícitos pero memoriza los datos de entrenamiento para su clasificación. Durante la predicción, se asigna a la instancia la clase con el recuento más alto entre los K vecinos más cercanos. La elección de K determina la suavidad del límite de decisión, donde valores más pequeños conducen a límites más complejos.

Regresión Logística:

Aunque a menudo se asocia con problemas de regresión, la regresión logística es en realidad un modelo de clasificación binaria. Utiliza la función logística (también conocida como sigmoide) para transformar la salida en una probabilidad, que se interpreta como la probabilidad de que una observación pertenezca a una clase particular. Durante el entrenamiento, el modelo ajusta los coeficientes de los predictores para minimizar la función de costo, utilizando métodos como el descenso de gradiente.

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("Error Cuadrático Medio (MSE):", mse)
print("Coeficiente de Determinación (R²):", r2)
```
